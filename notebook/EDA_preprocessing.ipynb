```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns

# Set number of plots per row
plots_per_row = 3
num_plots = len(numeric_cols)
num_rows = -(-num_plots // plots_per_row)  # Ceiling division

fig, axes = plt.subplots(num_rows, plots_per_row, figsize=(18, num_rows * 4))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.histplot(df_clean[col], kde=True, bins=30, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')

# Hide any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
categorical_cols = df_clean.select_dtypes(include=["category"]).columns

for col in categorical_cols:
    plt.figure(figsize=(10,4))
    df_clean[col].value_counts().head(15).plot(kind="bar", color="steelblue")
    plt.title(f"Distribution: {col}")
    plt.tight_layout()
    plt.show()
```

```python
sns.scatterplot(data=df_clean, x="totalpremium", y="totalclaims", alpha=0.4)
plt.title("Total Premium vs Total Claims")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df_clean.select_dtypes(include="number").corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap â€” Numeric Variables")
plt.show()
```

```python
for col in ["totalclaims", "customvalueestimate", "suminsured"]:
    if col in df_clean.columns:
        plt.figure(figsize=(8,4))
        sns.boxplot(x=df_clean[col], color="red")
        plt.title(f"Outliers â€” {col}")
        plt.show()
```

```python
if "transactionmonth" in df_clean.columns:
    monthly = df_clean.groupby(df_clean["transactionmonth"].dt.to_period("M")).agg({
        "totalclaims": "sum",
        "totalpremium": "sum"
    })

    monthly.index = monthly.index.to_timestamp()

    plt.figure(figsize=(12,6))
    plt.plot(monthly.index, monthly["totalclaims"], label="Total Claims", linewidth=2)
    plt.plot(monthly.index, monthly["totalpremium"], label="Total Premium", linewidth=2)
    plt.title("Monthly Trend â€” Claims vs Premium")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
bubble = df_clean.groupby("province").agg({
    "lossratio": "mean",
    "totalclaims": "sum",
    "totalpremium": "sum"
})

plt.figure(figsize=(12,7))
plt.scatter(
    bubble["lossratio"],
    bubble["totalclaims"],
    s=bubble["totalpremium"] / 500,  # bubble size
    alpha=0.6,
    c=bubble["lossratio"],
    cmap="viridis"
)
plt.colorbar(label="Loss Ratio")
plt.xlabel("Loss Ratio (Avg)")
plt.ylabel("Total Claims (Sum)")
plt.title("Province Risk Bubble Map")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get top 10 most frequent makes
top_makes = df_clean["make"].value_counts().head(10).index
avg_premium_by_make = df_clean[df_clean["make"].isin(top_makes)] \
    .groupby('make')['totalpremium'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_premium_by_make.index, y=avg_premium_by_make.values, palette='viridis')
plt.title('Average Total Premium by Top 10 Auto Makes')
plt.ylabel('Average Premium')
plt.xlabel('Auto Make')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by PostalCode and calculate mean of numerical values
zipcode_agg = df_clean.groupby("postalcode")[["totalpremium", "totalclaims"]].mean()

# Reset index to make PostalCode a column
zipcode_agg = zipcode_agg.reset_index()

# Compute the correlation matrix
corr_matrix = zipcode_agg.corr()

# Create a diverging color palette
cmap = sns.diverging_palette(220, 20, as_cmap=True)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data=corr_matrix, cmap=cmap, annot=True, fmt=".2f", square=True, linewidths=0.5)
plt.title("Correlation Matrix for TotalPremium and TotalClaims by PostalCode")
plt.show()
```
